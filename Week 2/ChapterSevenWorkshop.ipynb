The built-in function **eval** takes a string and evaluates it using the Python interpreter. For example (from the Python Prompt):

```
import math

print(eval('1 + 2 * 3'))
print(eval('math.sqrt(5)'))
print(eval('type(math.pi)'))

```
Write a function called **eval_loop** that iteratively prompts the user, takes the resulting input and evaluates it using eval, and then prints the result. It should continue until the user enters the string 'done'.


````python
# Exercise 7.1
def eval_loop():
    
    result = None # variable to handle the result
    while True:  # while true means this always runs
        userInput = input("please enter your expression or 'done' to quit: ")  # provide instruction
        if userInput.lower() == "done": # convert the input to lowercase and check if it is "done"
            break # if so, break out the loop
        result = eval(userInput) # do the evaluation and store result
        print(result) # print the result
    
eval_loop() # initiate program


````

Use the function is_leap(y) developed in an earlier exercise to calculate the number of leap years that occur between any two years (inclusive) entered by the user.

````python
# Exercise 7.2
"""
Function to determine if a given year is a leap year.
year : the year we wish to check if is a leap year.
"""
def is_leap(year):   
    result = False
    if year % 4 == 0:
        result = True
    elif year % 100 == 0:
        result = False
    elif year % 400 == 0:
        result = True
    return result

# get the years we are interested in
# We have yet to cover any tuples or dicts, so have chosen to work out the answer without using these.
# If dicts, lists, tuples were an optionm, this logic could have been included in a single function
year1 = input("Enter first year: ")
year2 = input("Enter second year: ")

# convert entered values to ints
year1 = int(year1)
year2 = int(year2)

# handle folk being awkward.
if year2 < year1:
    print("Year 2 must be greater than Year 1.")
    exit

# we can use a for loop to loop through the years.  However the range function is not inclusive, so we add 1 there to cover that
noOfLeaps = 0;
for x in range(year1, year2+1):
    res = is_leap(x)

    # if res (ie True), then we have a leap year.  Inform the user, add to a counter
    if res:
        print(x," is a leap year.")
        noOfLeaps += 1
    else:
        print(x," is not a leap year.")

# outside the for loop now, we have finished our check.  Display the results to the user.
print("In the range", year1, "to", year2,"there are", noOfLeaps,"leap years")
    
# NOTES: using dicts it could be done something list below - this keeps the is_leap function and wraps it in another function which handles all the user checking
#""" is_leap takes a single year and returns True or False if the year is a leap year or not """
#def is_leap(year):   
#    result = False
#    if year % 4 == 0:
#        result = True
#    elif year % 100 == 0:
#        result = False
#    elif year % 400 == 0:
#       result = True
#    return result

#""" getLeapYearForRange takes 2 years and adds these to a dictionary.  The format is of the dictionary is year: True /False """
#def getLeapYearForRange(year1, year2):
#
#    # create the dictionary    
#    leaps = {}
#
#    # ensure we are working with ints
#    year1 = int(year1)
#    year2 = int(year2)
#
#    # ensure that we have year 2 GREATER than year 1
#    if year2 < year1:
#        print("this is not valid.")
#        exit
#    
#    # for loop to call the is_leap function and storing the result in a dictionary
#    for x in range(year1, year2+1):
#        res = is_leap(x)
#
#        if res:
#            print(x," is a leap year.")
#            leaps[x] = res 
#        else:
#            print(x," is not a leap year.")           
#    return leaps
#
## ask for the years
#year1 = input("Enter first year: ")
#year2 = input("Enter second year: ")
#
## call the getLeapYearForRange function, passing in the years entered.
#countOfLeaps = getLeapYearForRange(year1, year2)
#
## print out to the user how many leap years we have
#print("There are ", len(countOfLeaps), "leap years in your range.")
````
